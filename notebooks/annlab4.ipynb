```python
import numpy as np
import matplotlib.pyplot as plt

def linear(x):
  return x

x = np.linspace(-5, 5, 100)  # Generate 100 evenly spaced points between -5 and 5 # Move this line outside the function

plt.figure(figsize=(15, 15))

plt.subplot(4, 2, 1)
plt.plot(x, linear(x))
plt.title('Linear')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
  return np.maximum(0, x)
  x = np.linspace(-5, 5, 100)  # Generate 100 evenly spaced points between -5 and 5

plt.figure(figsize=(15, 15))
plt.subplot(4, 2, 2)
plt.plot(x, relu(x))
plt.title('ReLU')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.01):
  return np.maximum(alpha * x, x)
  x = np.linspace(-5, 5, 100)  # Generate 100 evenly spaced points between -5 and 5

plt.figure(figsize=(15, 15))

plt.subplot(4, 2, 3)
plt.plot(x, leaky_relu(x))
plt.title('Leaky ReLU')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
  return 1 / (1 + np.exp(-x))
  x = np.linspace(-5, 5, 100)  # Generate 100 evenly spaced points between -5 and 5

plt.figure(figsize=(15, 15))

plt.subplot(4, 2, 4)
plt.plot(x, sigmoid(x))
plt.title('Sigmoid')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def hyperbolic_tan(x):
  return np.tanh(x)
  x = np.linspace(-5, 5, 100)  # Generate 100 evenly spaced points between -5 and 5

plt.figure(figsize=(15, 15))

plt.subplot(4, 2, 5)
plt.plot(x, hyperbolic_tan(x))
plt.title('Hyperbolic Tan')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def bipolar_sigmoid(x):
  return (1 - np.exp(-x)) / (1 + np.exp(-x))
  x = np.linspace(-5, 5, 100)  # Generate 100 evenly spaced points between -5 and 5

plt.figure(figsize=(15, 15))

plt.subplot(4, 2, 6)
plt.plot(x, bipolar_sigmoid(x))
plt.title('Bipolar Sigmoid')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def softplus(x):
  return np.log(1 + np.exp(x))
  x = np.linspace(-5, 5, 100)  # Generate 100 evenly spaced points between -5 and 5

plt.figure(figsize=(15, 15))

plt.subplot(4, 2, 7)
plt.plot(x, softplus(x))
plt.title('Softplus')
```
